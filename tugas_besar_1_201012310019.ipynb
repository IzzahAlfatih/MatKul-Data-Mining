# EDA

## Combine Data

```python
import pandas as pd
import glob

# List all CSV files in the directory
file_list = glob.glob('/content/PPG_Signal/*.csv')

# Initialize an empty list to store DataFrames
dfs = []

# Read each CSV file and append it to the list
for file in file_list:
    df = pd.read_csv(file, header=0)
    dfs.append(df)

# Concatenate all DataFrames in the list along the rows axis
combined_df = pd.concat(dfs, ignore_index=True)

# Display the combined DataFrame
combined_df
```

## Check for NaN and 0 value in the data. Fill with average.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Assuming combined_df is your DataFrame after combining CSV files
# Check for NaN values before filling
nan_values_before = combined_df.isnull().sum().sum()

# Fill NaN values with the column average
filled_df = combined_df.fillna(combined_df.mean())

# Check for NaN values after filling
nan_values_after = filled_df.isnull().sum().sum()

# Check for zero values before and after filling NaN
zero_values_before = (filled_df == 0).sum().sum()

# Fill zero values with the column average
filled_df.replace(0, filled_df.mean(), inplace=True)

# Check for zero values after filling
zero_values_after = (filled_df == 0).sum().sum()

# Display total NaN and zero values before and after filling
print("Total NaN values before filling:", nan_values_before)
print("Total NaN values after filling:", nan_values_after)
print("Total Zero values before filling:", zero_values_before)
print("Total Zero values after filling:", zero_values_after)
```

## Selecting highly correlated data only ( more than 0.5 correlation)

```python
# Extract features (X) and target values (y)
X = filled_df.iloc[:, :4200]  # Features are in columns 0 to 4199
y = filled_df.iloc[:, 4200:4202]  # Target values are in columns 4200 and 4201

# Compute the correlation matrix
corr_matrix = X.corrwith(y.iloc[:, 0])  # Compute correlation with the first output column
corr_matrix = corr_matrix.abs().sort_values(ascending=False)  # Sort by absolute correlation values

selected_features = corr_matrix[corr_matrix.abs() > 0.50].index.tolist()

# Select only the features with correlation > 0.50
X_selected = X[selected_features]

print("Selected Features:", selected_features)
```

## Splitting data

```python
from sklearn.model_selection import train_test_split

# Split the selected data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.15, random_state=42)
```

# Model Training

Since the output is continuous, I assume this is a regression matter.

## Linear Regression (Scikit-Learn Regression Model)

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Create and train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the testing data
y_test_pred = model.predict(X_test)
y_train_pred = model.predict(X_train)

# Evaluate the model performance
train_mse = mean_squared_error(y_train, y_train_pred)
test_mse = mean_squared_error(y_test, y_test_pred)

print("Training Mean Squared Error:", train_mse)
print("Testing Mean Squared Error:", test_mse)
```
