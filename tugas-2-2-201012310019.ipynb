```python
# Dataset Directory

customers = '/content/dataset-data-understanding-tugas-2/customers.csv'
gas = '/content/dataset-data-understanding-tugas-2/gasstations.csv'
products = '/content/dataset-data-understanding-tugas-2/products.csv'
transactions = '/content/dataset-data-understanding-tugas-2/transactions_1k.csv'
yearmonth = '/content/dataset-data-understanding-tugas-2/yearmonth.csv'

df_customers = pd.read_csv(customers)
df_gas = pd.read_csv(gas)
df_products = pd.read_csv(products)
df_transactions = pd.read_csv(transactions)
df_yearmonth = pd.read_csv(yearmonth)
```

```python
df_transactions = pd.read_csv(transactions)
df_transactions
```

### 1. Buatlah table list Top 5 Customers yang memiliki nilai transaksi paling banyak menggunakan dataframe pada python !

```python
customer_transactions = df_transactions.groupby('CustomerID')['Price'].sum().reset_index()

# Sort customers by total transaction value in descending order
customer_transactions_sorted = customer_transactions.sort_values(by='Price', ascending=False)

# Filter the top 5 customers
top_5_customers = customer_transactions_sorted.head(5)

# Merge dataframes on 'GasStationID'
merged_df_customers = pd.merge(top_5_customers, df_customers, on='CustomerID')

# Print the merged dataframe
merged_df_customers
```

### 2. Buatlah list Top 5 Gas stations yang memiliki nilai transaksi paling banyak menggunakan dataframe pada python!

```python
import pandas as pd

# Assuming your dataframe is named df
# Group by GasStation and calculate total transaction value
gas_station_transactions = df_transactions.groupby('GasStationID')['Price'].sum().reset_index()

# Sort gas stations by total transaction value in descending order
gas_station_transactions_sorted = gas_station_transactions.sort_values(by='Price', ascending=False)

# Filter the top 5 gas stations
top_5_gas_stations = gas_station_transactions_sorted.head(5)

# Merge dataframes on 'GasStationID'
merged_df_gas_stations = pd.merge(top_5_gas_stations, df_gas, on='GasStationID')

# Print the merged dataframe
merged_df_gas_stations
```

### 3. Buatlah list Top 5 Jenis produk yang memiliki nilai transaksi paling banyak menggunakan dataframe pada python!

```python
# Assuming your dataframe is named df
# Group by ProductID and calculate total transaction value
product_transactions = df_transactions.groupby('ProductID')['Price'].sum().reset_index()

# Sort products by total transaction value in descending order
product_transactions_sorted = product_transactions.sort_values(by='Price', ascending=False)

# Filter the top 5 products
top_5_products = product_transactions_sorted.head(5)

merged_df_product = pd.merge(top_5_products, df_products, on='ProductID')

merged_df_product
```

### 5. Carilah waktu terbaik (hari dan jam) dimana paling banyak user gas station melakukan transaksi!

```python
# Group by 'Date' and 'Time', and count the number of transactions
transactions_count = df_sorted.groupby(['Date', 'Time']).size().reset_index(name='TransactionCount')

# Find the row with the maximum number of transactions
max_transactions_row = transactions_count.loc[transactions_count['TransactionCount'].idxmax()]

# Extract the best date and time with the most transactions
best_date = max_transactions_row['Date'].date()  # Extract only the date part
best_time = max_transactions_row['Time']
max_transactions = max_transactions_row['TransactionCount']

# Print the result
print(f"The best date and time with the most transactions: Date={best_date}, Time={best_time}, Transactions={max_transactions}")
```
