```python
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_excel('/content/data1.xlsx')
data1
```

```python
data2 = pd.read_excel('/content/data2.xlsx')
data2
```

```python
data2 = data2.drop(data2.columns[0],axis=1)
data2
```

```python
result_df = pd.concat([data1, data2], axis=1)
result_df
```

```python
result_df['annual_bonus'] = result_df.apply(lambda row: row[['penjualan_kuartal1', 'penjualan_kuartal2', 'penjualan_kuartal3', 'penjualan_kuartal4']].sum() * 350000
                                if row['lama_kerja(tahun)'] >= 4
                                else row[['penjualan_kuartal1', 'penjualan_kuartal2', 'penjualan_kuartal3', 'penjualan_kuartal4']].sum() * 200000,
                                axis=1)
```
